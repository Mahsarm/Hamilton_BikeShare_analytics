reading historical weather data:

The data is collected from http://climate.weather.gc.ca/

```python
import numpy as np
import pandas as pd
import json
```

```python
weather_data = pd.read_csv('data/weather.csv', sep=';')
columns = weather_data.iloc[[13]]
type(columns.values)
columns = columns.values[0][0].replace('"', ' ').split(",")
df = pd.DataFrame([ weather_data.iloc[i].values[0].replace('"', ' ').split(",") for i in range(14,15) ],columns=columns)
df
#len(weather_data.iloc[14].values[0].split(","))
```

```python
data = [weather_data.iloc[i].values[0].replace('"', ' ').split(",") for i in range(14,len(weather_data)) ]
len(data)
```

Cleaning the weather data: some lines have more than 25 columns:

trying to see what those lines look like:

```python
s1 = [x for x in data if len(x) == 27] 
df1 = pd.DataFrame(s1)
df1
```

```python
df1[24] = df1[[24, 25, 26]].apply(lambda x: ', '.join(x), axis=1)
df1 = df1.drop([25,26], axis=1)
df1.head()
```

```python
s2 = [x for x in data if len(x) == 26] 
df2 = pd.DataFrame(s2)
df2.head()
```

```python
df2[24] = df2[[24, 25]].apply(lambda x: ', '.join(x), axis=1)
df2 = df2.drop(25, 1)
df2.head()
```

```python
s3 = [x for x in data if len(x) == 25] 
df3 = pd.DataFrame(s3)
df3.head()
```

```python
weather_data = (df1.append([df2, df3])).sort_values(by=0)
weather_data.columns = columns
weather_data
```

```python
weather_data.to_csv('data/weather-data.csv')
```
